```python
# Select 1100 emails of several authors
# First exclude the last two (Taylor and Martin) since htey only have 3 emails each
exec_authors = df_exec['Sender'].unique()[0:-2]
# Choose 4 authors and 300 emails from each
selected_execs = np.random.choice(exec_authors, size=4, replace=False)
# print("We choose 300 emails each from", selected_execs)

# The list of indices of selected emails
index_list = []
for exec in selected_execs:
  # Find all emails for the given selected exec
  index_list_exec = df_exec.index[(df_exec['Sender'] == exec)].tolist()
  # Select 300 of those emails randomly
  index_short_list = random.sample(index_list_exec, 300)
  index_list += index_short_list
# Make a data frame containing only the 1200 emails selected.
#df_exec_selected = pd.DataFrame()
df_exec_selected = df_exec.loc[df_exec.index[index_list]]
df_exec_selected = df_exec_selected.reset_index()
df_exec_selected
```

**Compare sentiment for execs and poi's**

```python
poi_positive_count = 0
poi_negative_count = 0
exec_positive_count = 0
exec_negative_count = 0

for sentiment in df['Sentiment']:
  if sentiment == 'POSITIVE':
    poi_positive_count += 1
  else:
    poi_negative_count += 1

for sentiment in df_exec_selected['Sentiment']:
  if sentiment == 'POSITIVE':
    exec_positive_count += 1
  else:
    exec_negative_count += 1

print("Positive sentiment in poi:", poi_positive_count)
print("Negative sentiment in poi:", poi_negative_count)
print("Positive sentiment in exec:", exec_positive_count)
print("Negative sentiment in exec:", exec_negative_count)

print("Proportion of positive sentiment in poi:", poi_positive_count/1109)
print("Proportion of positive sentiment in exec:", exec_positive_count/1200)
```
